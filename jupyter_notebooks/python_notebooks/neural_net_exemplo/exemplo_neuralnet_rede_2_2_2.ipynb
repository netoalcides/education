## Carrega pacotes

```python
#%matplotlib inline
#import numpy as np
#import matplotlib.pyplot as plt
#import pandas as pd

import lib.nn as nn
```

## Gera rede neural do primeiro exemplo

```python
# learning = 0.5
neural = nn.NeuralNetwork(2, 2, 2, 
                          hidden_layer_weights=[0.15, 0.2, 0.25, 0.3], 
                          hidden_layer_bias=0.35, 
                          output_layer_weights=[0.4, 0.45, 0.5, 0.55], 
                          output_layer_bias=0.6)
neural.inspect()
```

## Resultados antes do treinamento

```python
print( 'MSE: {}'.format( round( neural.calculate_total_error( [[[0.05, 0.1], [0.01, 0.99]]] ), 4 ) ) )

print( 'H_1: {}'.format( round( neural.hidden_layer.get_outputs()[0], 4) ) )
print( 'H_2: {}'.format( round( neural.hidden_layer.get_outputs()[1], 4) ) )

print( 'O_1: {}'.format( round( neural.output_layer.get_outputs()[0], 4) ) )
print( 'O_2: {}'.format( round( neural.output_layer.get_outputs()[1], 4) ) )
```

## Treina a rede

```python
epocas = 500
for i in range(epocas):
    neural.train([0.05, 0.1], [0.01, 0.99])
    print( 'Época: {}/{} MSE: {}'.format(i, epocas-1, 
                                      round( neural.calculate_total_error( [[[0.05, 0.1], [0.01, 0.99]]] ), 4 ) ), end = '\r' )
```
